```python
grouped_data = p1.groupby(['Model','VM']).agg({
                                                'CPU_MAE':'mean',
                                                'CPU_RMSE':'mean',
                                                'MPU_MAE' : 'mean',
                                                'MPU_RMSE' : 'mean',
                                                'CPU cores':'max',
                                                'CPU capacity provisioned [MHZ]':'mean',
                                                'predicted_cpu_usage [%]':'max',
                                                'CPU usage [MHZ]' : 'max',
                                                'Memory capacity provisioned [KB]' : 'mean',
                                                'predicted_memory_usage [%]':'max',
                                                'Memory usage [KB]':'max'
                                            
})
```

## Taking only Memory into Consideration, Filtering all memory less than 8 Gb


## For VM 3 Highest predicted memory usage % is 10.515 and cpu % is 0.9, Hence we are using less that

```python
VM_3=cost_df[(cost_df['Memory']<8) & (cost_df['Clock Speed(GHz)']<2.5) & (cost_df['Cores']<2)]
VM_3.loc[:,'VM']=3
VM_3.loc[:,'Cost for Linux On Demand for 1 Year']=VM_3['Linux On Demand cost']*8640
VM_3.loc[:,'Cost for Linux Reserved for 1 Year']=VM_3['Linux Reserved cost']*8640
```

## For VM 198 Highest predicted memory usage % is 17 and cpu % is 50, Hence we are using less than or equal

```python
VM_198=cost_df[(cost_df['Memory']<=8) & (cost_df['Clock Speed(GHz)']<=2.5) & (cost_df['Cores']==2)]
VM_198.loc[:,'VM']=198
VM_198.loc[:,'Cost for Linux On Demand for 1 Year']=VM_198['Linux On Demand cost']*8640
VM_198.loc[:,'Cost for Linux Reserved for 1 Year']=VM_198['Linux Reserved cost']*8640
```

## For VM 1 Highest predicted memory usage % is 13 and cpu % is 9, Hence we are using less than or equal for 1 year

```python
VM_1=cost_df[(cost_df['Memory']<=8) & (cost_df['Clock Speed(GHz)']<2.5) & (cost_df['Cores']<=2)]
VM_1.loc[:,'VM']=1
VM_1.loc[:,'Cost for Linux On Demand for 1 Year']=VM_1['Linux On Demand cost']*8640
VM_1.loc[:,'Cost for Linux Reserved for 1 Year']=VM_1['Linux Reserved cost']*8640
```

```python
VM_1
```

```python
frames = [VM_1, VM_3, VM_198]

recommendations = pd.concat(frames)
```

```python
recommendations_1=recommendations[recommendations['VM']==1]
recommendations_3=recommendations[recommendations['VM']==3]
recommendations_198=recommendations[recommendations['VM']==198]
recommendations_198
```

```python
recommendations_1['Cost Saved on Linux Demand']=forecasted_grouped_data[forecasted_grouped_data['VM']==1].loc[:,'Forecasted Cost of OnDemand Linux Service for 1 Year'].values-recommendations_1.loc[:,'Cost for Linux On Demand for 1 Year'].values
recommendations_1['Cost Saved on Linux Reserved']=forecasted_grouped_data[forecasted_grouped_data['VM']==1].loc[:,'Forecasted Cost of Reserved Linux Service for 1 Year'].values-recommendations_1.loc[:,'Cost for Linux Reserved for 1 Year'].values
```

```python
print(forecasted_grouped_data[forecasted_grouped_data['VM']==1].loc[:,'Forecasted Cost of OnDemand Linux Service for 1 Year'].values.shape)
print(recommendations_1.loc[:,'Cost for Linux On Demand for 1 Year'].values.shape)
```

```python
recommendations_3['Cost Saved on Linux Demand']=forecasted_grouped_data[forecasted_grouped_data['VM']==3].loc[:,'Forecasted Cost of OnDemand Linux Service for 1 Year'].values-recommendations_3.loc[:,'Cost for Linux On Demand for 1 Year'].values
recommendations_3['Cost Saved on Linux Reserved']=forecasted_grouped_data[forecasted_grouped_data['VM']==3].loc[:,'Forecasted Cost of Reserved Linux Service for 1 Year'].values-recommendations_3.loc[:,'Cost for Linux Reserved for 1 Year'].values
```

```python
recommendations_198['Cost Saved on Linux Demand']=forecasted_grouped_data[forecasted_grouped_data['VM']==198].loc[:,'Forecasted Cost of OnDemand Linux Service for 1 Year'].values[0]-recommendations_198.loc[:,'Cost for Linux On Demand for 1 Year'].values
recommendations_198['Cost Saved on Linux Reserved']=forecasted_grouped_data[forecasted_grouped_data['VM']==198].loc[:,'Forecasted Cost of Reserved Linux Service for 1 Year'].values[0]-recommendations_198.loc[:,'Cost for Linux Reserved for 1 Year'].values
```

```python

frames_rec = [recommendations_1, recommendations_3, recommendations_198]

recommendations_final = pd.concat(frames_rec)
```
